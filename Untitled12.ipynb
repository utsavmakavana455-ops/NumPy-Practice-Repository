```python
import numpy as np

# 2-D array of product prices
prices = np.array([
    [100, 200, 150, 300],
    [120, 180, 160, 280],
    [90, 220, 140, 310],
    [110, 210, 170, 290]
])

# 1-D array of tax rates
tax_rates = np.array([0.05, 0.10, 0.08, 0.12])

# Apply tax rates using broadcasting
final_prices = prices * (1 + tax_rates)

# Compute total price for each store
store_totals = np.sum(final_prices, axis=1)

# Find the store with the highest total
highest_store = np.argmax(store_totals)

# Display results
print("Original Prices:")
print(prices)

print("\nFinal Prices (After Tax):")
print(final_prices)

print("\nTotal Price per Store:")
print(store_totals)

print("\nStore with the Highest Total:", highest_store)
print("Highest Total Price:", store_totals[highest_store])
```
